#### Exercise 5: Distances *(4 points)*

In this exercise, we want to calculate the distances from Bern to other cities in the world.

Source: [https://simplemaps.com/data/world-cities](https://simplemaps.com/data/world-cities)

1. Load the `data/csv/worldcities.csv` using pandas and store it in the variable `cities_df`.


```python
# TODO
cities_df = pd.read_csv('../data/csv/worldcities.csv')
```

2. For the next task, we need the latitude and longitude in radians instead of degrees. Create new columns `lat_rad` and `lng_rad` with the converted value in radians.

Hint: Use the suitable NumPy Ufunc.

```python
# TODO

import numpy as np
cities_df['lat_rad'] = np.radians(cities_df['lat'])
cities_df['lng_rad'] = np.radians(cities_df['lng'])

cities_df.head()
```

2. Calculate the distance between Bern and all the cities in a new column `distance_to_bern`. You can do this by using the Haversine formula:

$$ d = r \cdot 2 \cdot \arcsin \left( \sqrt{\sin^2 \left( \frac{lat2 - lat1}{2}\right) + \cos(lat1) \cdot \cos(lat2) \cdot \sin^2 \left( \frac{lng2 - lng1}{2}\right)} \right), $$

where $lat1$ and $lat2$ are the latitudes in radians, and $lng1$ and $lng2$ the longitudes in radians, and the $r$ the radius of the Earth. 

Hints:
- For Bern, use lat=46.94798°, lng=7.44743° and $r=6371 km$. The calculated distances will be in kilometers.
- Do not forget to convert the coordinates of Bern to radians.
- There are many suitable Ufuncs to use, e.g., `np.arcsin`.


```python
# TODO
r = 6371
lat_bern = np.radians(46.94798)
lng_bern = np.radians(7.44743)

lat2 = cities_df['lat_rad']
lng2 = cities_df['lng_rad']

# Haversine formula
dlat = cities_df['lat_rad'] - lat_bern
dlng = cities_df['lng_rad'] - lng_bern
a = np.sin(dlat / 2) ** 2 + np.cos(lat_bern) * np.cos(cities_df['lat_rad']) * np.sin(dlng / 2) ** 2
cities_df['distance_to_bern'] = r * 2 * np.arcsin(np.sqrt(a))

cities_df[['city', 'distance_to_bern']].head()
```
