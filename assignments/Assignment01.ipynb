#### Exercise 6: DRY *(3 points)*

An important software development principle is **DRY** (Don't Repeat Yourself).  
It aims to reduce duplicated code by centralizing logic and reusing functions where possible.

1. Create a single list of tuples containing the student name and the corresponding grade.

2. Implement a function `print_result` that displays the result for a single student.

3. Create a loop to iterate through the list of tuples and use the `print_result` function to print each result.

The final output must exactly match the current output.

````python
name_1 = "Anna"
grade_1 = 5.5
print(f"Student: {name_1}, Grade: {grade_1}, Passed: {grade_1 >= 4}")

name_2 = "Ben"
grade_2 = 6
print(f"Student: {name_2}, Grade: {grade_2}, Passed: {grade_2 >= 4}")

name_3 = "Clara"
grade_3 = 3.5
print(f"Student: {name_3}, Grade: {grade_3}, Passed: {grade_3 >= 4}")
````

````python
students = ({"name": "Anna", "grade": 5.5}, {"name": "Ben", "grade": 6}, {"name": "Clara", "grade": 3.5})


def print_result(student):
    print(f"Student: {student['name']}, Grade: {student['grade']}, Passed: {student['grade'] >= 4}")


for student in students:
    print_result(student)
````

#### Exercise 8: Loops *(2 points)*

Implement a function `average_word_length`, that accepts a list of strings and calculates the average length of all the words given as input.

Example: `print(average_word_length(["moon", "sun", "stars"]))` should output `4.0`

````python
def average_word_length(words: list):
    sum = 0
    for word in words:
        sum += len(word)
    return sum / len(words)


print(average_word_length(["moon", "sun", "stars"]))
````
