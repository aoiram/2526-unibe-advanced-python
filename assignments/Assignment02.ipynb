#### Exercise 1: Lists *(3 points)*

````python
from statistics import mean

from numpy.ma.core import sqrt

list_1 = [1, 2, 3]
list_2 = list_1

list_2.append(4)

print("List 1:", list_1)
print("List 2:", list_2)
````

Explain why both lists contain 4, even though we do not have code directly involving `list_1`.

**Your answer**: Appending is a mutation operation. list_1 and list_2 are references to the same object. When you mutate the object, all references will show the updated object. In this case, with the appended 4.

Demonstrate your explanation by using `id(...)` and comments.

````python
print(f'Are ID of list_1 and list_2 equals : {id(list_1) == id(list_2)}')
# id retrieves the integer value of the memory address. When we compare both list_1 and list_2's id, we find that they are the same value hence showing that they are references to the same object
````

Explain whether this phenomenon also occurs with immutable types.

**Your answer**: This does not happen with immutable types

Say you do \
a = 10 \
b = a \
a = 5

This is because when you create a pointer to the object, say \"a\" from above, and then you reassign it, (as it can only be reassigned and not changed: it is immutable), it will not affect the value of another pointer, because b is still pointing to the original object. So here, b == 10.

#### Exercise 2: Filtering *(2 points)*

Implement a function `odd_numbers` that accepts a list of integers and returns a list containing only the odd numbers of this list.

Example: `print(odd_numbers([1, 2, 3]))` should output `[1, 3]`

````python
def odd_numbers(list):
    out = []
    for number in list:
        # if number % 2 == 1:   # Rohit version
        if number & 1 == 1:  # Mario version : bitwise comparison is faster than modulo
            out.append(number)
    return out


print(odd_numbers([1, 2, 3, 4, 5]))
````

#### Exercise 4: Lottery *(5 points)*

- In this exercise, we use the `randint` function, which can be used to generate pseudorandom integers.
- Implement a function `gamble` that accepts an inputs `fee`.
- The function should initialize a variable `spending` with 0.
- Then, the `fee` should be deducted from `spending` and 3 random integers between 0 and 9 should be generated.
- If all three numbers are `6`, the program stops (victory).
- Otherwise, it should continue by deducting the fee again from `spending` and by generating another set of 3 random integers.
- Use `print` in every iteration to ouput the current state (the 3 random numbers and the current `spending`). In the end, the program should return the (negative) `spending`.

````python
import random


def gamble(fee, verbose=False):
    spending = 0

    while True:
        spending -= fee
        nums = [random.randint(1, 9) for _ in range(3)]

        if verbose:
            print(f"Numbers: {nums} , Spent: {spending}")

        if nums == [6, 6, 6]:
            break
    return spending


gamble(2, True)
````

- Create a loop and call the function `gamble` many times (e.g., 1000) with a fee of 5 CHF.
- Change the `gamble` function to make the many calls of `print` optional by adding an argument `verbose`. If `gamble(..., verbose=False`), there should be no prints.
- Add the return value of the `gamble` function to a list.
- Calculate the average spending.

As a casino owner, what is the maximum jackpot prize that you can award to winners for a win without making loss in the long run?

````python
import numpy as np

n = 1000
fee = 5

spendings = []
for i in range(n):
    spendings.append(gamble(fee))

mean = sum(spendings) / n
print(f'Average spending of {mean} on {n} runs with fee = {fee}')
````

While short-term deviations in spending may result in temporary profits or losses, these fluctuations will eventually average out to a zero balance over time.

#### Exercise 5: Number divisors *(4 points)*

A divisor of a number, is another number that divides the first number without reminder.

Example: 4 is a divisor of 20. 1 and the number itself are always divisors. 

Write a function `find_divisors` that takes a number in input and returns *all* its divisors.

Example:
```python 
find_divisors(20)  # returns [1, 2, 4, 5, 10, 20]
find_divisors(7)  # returns [1, 7]
```

````python
from math import sqrt


def find_divisors(number):
    divisors = set()  # Using set so it sort it nicely
    # Ranging from 1 to sqrt bc we know that bigger divisors would result in a number bigger than the number, other half is covered by the corresponding divisor
    for i in range(1, int(sqrt(number)) + 1):
        if number % i == 0:
            divisors.add(i)
            corresponding_divisor = number // i
            if i != corresponding_divisor:  # Add the corresponding divisors if not the same
                divisors.add(corresponding_divisor)
    return divisors


print(find_divisors(20))
````

Implement a function `find_common_divisors` that accepts to integers and returns a sorted list of common divisors (high divisors first).

Hints: Use the function `find_divisors` from above. Sets can be helpful to implement list intersections.

````python
def find_common_divisors(a, b):
    return set(find_divisors(a)) & set(find_divisors(b))

find_common_divisors(56, 48)
````
